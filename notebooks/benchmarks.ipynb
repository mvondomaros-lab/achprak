```python
import achprak
import timeit
import contextlib
import os
```

```python
def new_template():
    return achprak.azobenzene.Template()


def run_sp():
    template = new_template()
    properties = achprak.azobenzene.Properties(template.atoms)
    properties.energy()


def run_opt_min():
    with open(os.devnull, "w") as devnull:
        with contextlib.redirect_stdout(devnull):
            template = new_template()
            opt = achprak.optimization.OptMin(template.atoms)
            opt.run()


def run_opt_ts():
    with open(os.devnull, "w") as devnull:
        with contextlib.redirect_stdout(devnull):
            template = new_template()
            opt = achprak.optimization.OptTS(template.atoms)
            opt.run()


def run_uv_vis():
    with open(os.devnull, "w") as devnull:
        with contextlib.redirect_stdout(devnull):
            template = new_template()
            uv_vis = achprak.uvvis.UVVis(template.atoms)
            uv_vis.calculate()
```

```python
%%timeit
run_sp()
```

```python
%%timeit
run_opt_min()
```

```python
%%timeit
run_opt_ts()
```

```python
%%timeit
run_uv_vis()
```
